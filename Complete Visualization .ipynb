```python
#load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as ps
import seaborn as sns
```

# Matplotlib

```python
#input data 
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
x,y,z  # we print all values 
```

```python
#simple plot
plt.figure(figsize=(10,6))
plt.plot(x,y,"r.",x,z,"go")
plt.title("sine-cosine curve")
plt.xlabel("x-Axis")
plt.ylabel("y-Axis")
plt.legend()
plt.show()
```

```python
plt.plot(y,z)
```

# Statistical  Data Visualization With seaborn

```python
df=sns.load_dataset('tips')     
df
```

```python
df1=sns.load_dataset("titanic")
df1
```

```python
df1.isnull().sum()   # null value count
```

```python
a=df1.isnull().sum()/df1.shape[0]*100 #null values percentage
a
```

```python
df1_drop=a[a>50].keys()     # dropable keys
```

```python
df1_drop
```

```python
df1=df1.drop(columns=df1_drop)
df1
```

```python
df1.isnull().sum()
```

```python
df1_null_clr=df1.dropna().isnull().sum().sum() 
df1_null_clr
```

```python
df3=df1.dropna()
```

```python
df3   # clean data of titanic
```

```python
df4=sns.load_dataset("iris")
df4.isnull().sum().sum()
```

```python
df4     # iris data set
```

# Scatter plot

```python
#scatterplot
plt.figure(figsize=(10,6))
sns.stripplot(x="species",y="petal_length",data=iris)
```

```python
#swarmplot 
plt.figure(figsize=(10,6))
sns.swarmplot(x="species",y="petal_length",data=iris)
```

# Bar chart

```python
plt.figure(figsize=(20,10))
sns.barplot(x="sex",y="survived",hue="class",data=titanic)
```

# Count plot

```python
plt.figure(figsize=(10,6))
sns.countplot(x="sex",data=titanic)
```

# Point plot

```python
plt.figure(figsize=(10,6))
sns.pointplot(x="class",y="survived",hue="sex",data=titanic, palette={'male':'g','female':'m'},
              markers=["^","o"],linestyles=["-","--"])
```
